```python
import pandas as pd
import numpy as np

```

```python
df = pd.read_csv("train.csv")
```

```python
df.head()
```

```python
texts = df["text"]
labels = df["class_label"]
#texts.shape
labels.shape
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf_vectorizer = TfidfVectorizer(analyzer='word', tokenizer= lambda x: x.split(), stop_words='english')
vectorized_texts = tfidf_vectorizer.fit_transform(texts)
```

```python
input_size = vectorized_texts.shape
for vec in vectorized_texts:
    print(vec.shape)
    
```

Base model

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score

clf = LogisticRegression()
cv_scores = cross_val_score(clf, vectorized_texts, labels, cv = 5)
print("Accuracy: {:.3f}".format(np.mean(cv_scores)))
```

MLP: 

```python
from keras.layers import Dense
from keras.models import Sequential

def ann_model():
    model = Sequential()
    model.add(Dense(100, input_shape=(19221,), activation="sigmoid"))
    model.add(Dense(2, activation = "sigmoid"))
    model.add(Dense(1, activation = "sigmoid"))
    print(model.summary())

    model.compile(optimizer = "adam" , loss="binary_crossentropy", metrics=['accuracy'])
    return model
    
```

```python
from sklearn.model_selection import train_test_split
from keras.utils import to_categorical

X_train, X_test, y_train, y_test = train_test_split(vectorized_texts, labels, test_size = 0.2, random_state=7) 
# y_train = to_categorical(y_train, 2)
# y_test = to_categorical(y_test, 2)
model = ann_model()
model.fit(X_train, y_train, epochs = 5, batch_size=10)
test_score = model.evaluate(X_test, y_test)
train_score = model.evaluate(X_train, y_train)
print(train_score)
print(test_score)
```
